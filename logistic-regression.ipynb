```python
# we will show the impact of the salary and the department 
df_cat=df_reg.drop(['satisfaction_level','last_evaluation','average_montly_hours',
                    'number_project','time_spend_company', 'Work_accident' , 
                    'promotion_last_5years'],axis=1)
for i in df_cat.columns:
    plt.figure(figsize=(10,8))
    sns.countplot(data= df_cat, 
                  x= i,
                  hue='left',  
                  order = df_cat[i].value_counts().index)
    
    plt.show()
```

after checking the department bar plot we can see that it doesn't have a big impact on the employee retention thus we will ignore it 

so in our logistic regression model we will mainly use the folowing variables :

**Satisfaction Level** /
**Average Monthly Hours** /
**Promotion Last 5 Years** /
**Salary**

```python
df_reg.info()
```

```python
# we create our small dataframe 
df = df_reg[['satisfaction_level', 'average_montly_hours','promotion_last_5years','salary']]
df
```

```python
# we need to convert the salary values into numbers thus we will use dummy variable
salary_dum = pd.get_dummies(df.salary, prefix="salary")
df2 = pd.concat([df, salary_dum],axis='columns')
df2.drop('salary', axis='columns', inplace=True)
df2
```

```python
# we split our dataframe into dependent and independent
x = df2
y = df_reg['left']
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.2)
```

```python
# we create our model 
model = LogisticRegression()
# we fit our model 
model.fit(x_train, y_train)
```

```python
# make prediction 
model.predict(x_test)
```

```python
model.predict([[0.2,50,0,0,0,1]])
```

```python
# we calculate the accuracy of the model 
model.score(x_test, y_test)
```
